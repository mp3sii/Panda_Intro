```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# Setup figure size and DPI
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['figure.dpi'] = 150
```

```python
xvals = np.arange(0, 10, 0.1) # creates an array of points from start to end (not included) based on the step
xvals
```

```python
# We call the plot function with x as each observation on the xvals array
# and y the sinus of each observation of xvals array
plt.plot(xvals, np.sin(xvals))

# Set plot axis labels
plt.xlabel('x axis')
plt.ylabel('y axis') 
```

<https://matplotlib.org/api/pyplot_api#matplotlib.pyplot.plot>

```python
plt.plot(xvals, np.sin(xvals), 
    label='y = sin(x)', 
    color = 'red', 
    linestyle = '--')
plt.plot(xvals, np.cos(xvals), label='y = cos(x)', color = 'blue', linestyle = ':')

# Set plot axis labels
plt.xlabel('x')
plt.ylabel('y')

plt.legend(loc='best', fontsize = 'x-small') # Request legend to be displayed... 
plt.title('Sin and Cos of x')
# Set axis scale
plt.axis([0, 10, -1.5, 1.5]) # [x0, x1, y0, y1]
plt.show()
```

# Line and Scatter Plots
## Linestyles
Interesting arguments:
* **linestyle:** Setup the style from solid,  dotted, etc..
* **linewidth:** Setup the width in points
* **color:** Define the line color
* **alpha:** Set transparency
* **zorder:** Define the order of alignment of the different plots
* **marker:** Create markers for each desired observation
    * **markersize**
    * **markevery:** to mark every x observation

```python
idx = 0
for linestyle in ('-', '--', ':', '-.'):
    plt.plot(xvals, np.sin(xvals)+idx, linestyle = linestyle)
    idx +=1

# Set plot axis labels
plt.xlabel('x axis')
plt.ylabel('y axis') 
```

```python
# Using markers instead of lines
idx = 0
for marker in ['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h',
              'H', '+', 'x', 'D', 'd', '|', '_']:
    plt.plot(xvals, np.sin(xvals)+idx, linestyle = 'none', marker = marker, markersize = 5) 
#     markevery = 10 would mark every 10
#    markersize = 10 would mark every 10ar
    idx+=1
plt.axis([-0.05, 0.05, -0.5, idx+2])
```

Scatter()
